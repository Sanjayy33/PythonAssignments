<h1>Task 2</h1>

```python
def total_cart_amount(prices):
    total_price = 0
    for i in range(len(prices)):
        conv_flt = float(prices[i])
        total_price += conv_flt
    return round(total_price,2)
product_price = ['150.5','20.3','300.0','480.36']
print(total_cart_amount(product_price))
```

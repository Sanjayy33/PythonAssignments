<h1>Task 1</h1>

```python
def calculate_final_price(price, discount_rate):
    discount_amount = price * (discount_rate / 100)
    final_price = price - discount_amount
    return final_price
print(calculate_final_price(1000, 15))
print(calculate_final_price(500, 50))
```

<h1>Task 3</h1>

```python
def format_price(price, currency='INR'):
    if currency == 'INR':
        return f"₹{price}"
    elif currency == 'USD':
        return f"${price}"
    else:
        return f"{currency} {price}" 
print(format_price(500))
print(format_price(1200, 'INR'))
print(format_price(50, 'USD'))
```

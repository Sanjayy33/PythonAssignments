```python
def ipl_team_name(ipl_team):
    long_name = []
    for i in ipl_team:
        if len(i) > 6:
            long_name.append(i)
    return long_name
ipl_team = ["cskings","royalcb","mi","pbsk","delhicapital","royalrajsthan","lsg"]
print(ipl_team_name(ipl_team))
```

<h1>Task 5</h1>

```python
cart_prices = [499, 0, 799, 0, 350, 650, 200]
final_total = 0
for price in cart_prices:
    if price == 0:
        continue
    if final_total + price > 2000:
        break
    final_total += price
print(f"Final Cart Total: ₹{final_total}")
```

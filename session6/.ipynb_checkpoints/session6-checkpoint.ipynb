<h1>Task 5</h1>

```python
def  get_unique_artists(spotify_playlist1, spotify_playlist2):
    unique_artists = spotify_playlist1.union(spotify_playlist2)
    return unique_artists
spotify_playlist1 = {"geeta dutt","kishore kumar","mohmmad rafi","lata mangeshkar","asha bhosle"}
spotify_playlist2 = {"kishore kumar","mohmmad rafi","mukesh","hemant kumar","lata mangeshkar"}
print(get_unique_artists(spotify_playlist1, spotify_playlist2))
```
